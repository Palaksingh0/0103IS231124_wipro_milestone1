```python
import sys

def calculate_happiness():
    # Read the three strings from command line arguments
    if len(sys.argv) < 4:
        return

    string1 = sys.argv[1]
    string2 = sys.argv[2]
    numbers_given = sys.argv[3]

    # Convert hyphenated strings into sets for fast lookup
    # Sets are used because we only care if a number exists in the group
    liked = set(string1.split('-'))
    disliked = set(string2.split('-'))

    # The third string is the sequence we iterate through
    target_list = numbers_given.split('-')

    happiness = 0

    for num in target_list:
        if num in liked:
            happiness += 1
        elif num in disliked:
            happiness -= 1

    print(happiness)

if __name__ == "__main__":
    calculate_happiness()
```

```python
#q3
import sys

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Ensure 10 numbers are provided
if len(sys.argv) == 11:
    numbers = [int(arg) for arg in sys.argv[1:]]
    prime_sum = sum(num for num in numbers if is_prime(num))
    print(f"The sum of prime numbers is: {prime_sum}")
else:
    print("Please provide exactly 10 numbers.")
```
