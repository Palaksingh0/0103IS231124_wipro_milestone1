//dictionary project 1


```python
# 1. Create a dictionary with people and facts
people_facts = {
    "Jeff": "Is afraid of Dogs.",
    "David": "Plays the piano.",
    "Jason": "Can fly an airplane."
}

# Function to display the dictionary
def display_facts(data):
    for name, fact in data.items():
        print(f"{name}: {fact}")
    print() # New line for spacing

# 2. Display initial facts
display_facts(people_facts)

# 3. Change a fact about one person
people_facts["Jeff"] = "Is afraid of heights."

# 4. Add an additional person and fact
people_facts["Jill"] = "Can hula dance."

# 5. Display the updated list
display_facts(people_facts)
```

//project 3

```python
# Initialize the student records dictionary
student_marks = {
    "Krishna":,
    "Arjun":,
    "Malika": [52, 56, 60]
}

# Get user input for the student's name
query_name = input("Enter a name: ")

# Check if the name exists in the record
if query_name in student_marks:
    # Retrieve the list of marks
    marks = student_marks[query_name]

    # Calculate the average using the provided formula
    # (sum of marks) / (no of subjects)
    average = sum(marks) / len(marks)

    # Output the result
    print(f"Average percentage mark: {average:.0f}")
else:
    print("Student record not found.")
```

```python
#q1
def count_case(s):
    upper = sum(1 for c in s if c.isupper())
    lower = sum(1 for c in s if c.islower())
    print(f"Uppercase: {upper}, Lowercase: {lower}")

# Example: count_case("Hello World") -> Uppercase: 2, Lowercase: 8
```

```python
#q2
def is_palindrome(s):
    # Case-insensitive check
    clean_s = s.lower()
    return clean_s == clean_s[::-1]

# Example: is_palindrome("Radar") -> True
```

```python
#q3
def repeat_first_two(s):
    if len(s) < 2:
        return s
    n = len(s)
    first_two = s[:2]
    return first_two * n

# Example: "Wipro" -> "WiWiWiWiWi"
```

```python
#q4
def remove_x(s):
    if not s:
        return s

    # Handle the last character first to maintain index integrity
    if len(s) > 0 and s[-1] == 'x':
        s = s[:-1]
    if len(s) > 0 and s[0] == 'x':
        s = s[1:]

    return s

# Example: "xHix" -> "Hi"
```

```python
#q5
def repeat_last_n(s, n):
    if n == 0:
        return ""
    last_n = s[-n:]
    return last_n * n

# Example: "Wipro", 3 -> "propropro"
```
