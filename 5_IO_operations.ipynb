```python
import re

filename = input("Enter file name: ")

with open(filename, "r") as file:
    lines = file.readlines()

# Number of lines
line_count = len(lines)

# Count words
freq = {}

for line in lines:
    words = re.findall(r"[A-Za-z]+", line.lower())
    for word in words:
        freq[word] = freq.get(word, 0) + 1

# Maximum frequency word
place = max(freq, key=freq.get).capitalize()

# Meeting time
if line_count <= 12:
    time = f"{line_count} AM"
else:
    time = f"{line_count-12} PM"

print("Meeting time:", time)
print("Meeting place:", place, "Street")
```

```python
#q2
n = int(input("Enter the number of lines to read: "))
file_path = 'example.txt'

with open(file_path, 'r') as file:
    for i in range(n):
        line = file.readline()
        if not line:
            break
        print(line.strip())
```

```python
#q5
file_path = 'example.txt'

with open(file_path, 'r') as file:
    words = file.read().split()
    longest_word = max(words, key=len)

print(f"The longest word is: {longest_word}")
```

```python
#q6
target_word = input("Enter the word to count: ")
file_path = 'example.txt'

with open(file_path, 'r') as file:
    content = file.read().lower()
    words = content.split()
    count = words.count(target_word.lower())

print(f"The word '{target_word}' appears {count} times.")
```
