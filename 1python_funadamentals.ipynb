```python
#q6
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


num = int(input("Enter a number: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
#q8
def sum_of_digits(n):
    total = 0
    while n > 0:

        total += n % 10

        n //= 10
    return total


number = 1234
print(f"I/P: {number}")
print(f"O/P: {sum_of_digits(number)}")
```

```python
# #q9
def reverse_number(n):
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
    return reversed_num

# Test cases
num1 = 1234
num2 = 1004

print(f"I/P: {num1} -> O/P: {reverse_number(num1)}")
print(f"I/P: {num2} -> O/P: {reverse_number(num2)}")
```

```python
#q10
def is_palindrome(n):

    num_str = str(n)

    if num_str == num_str[::-1]:
        print(f"{n} is a palindrome.")
    else:
        print(f"{n} is not a palindrome.")


is_palindrome(110011)
is_palindrome(1234)
```
