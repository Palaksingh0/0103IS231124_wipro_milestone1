```python
try:
    filename = input("Enter the file name: ").strip()


    if not filename.endswith(".txt"):
        filename += ".txt"

    total_items = 0
    free_items = 0
    total_amount = 0
    discount = 0

    with open(filename, "r") as file:
        for line in file:
            line = line.strip()


            if line == "":
                continue

            parts = line.split()

            if len(parts) != 2:
                print("Invalid line skipped:", line)
                continue

            item = parts[0]
            value = parts[1]

            if item.lower() == "discount":
                try:
                    discount = float(value)
                except ValueError:
                    print("Invalid discount value.")
                continue

            if value.lower() == "free":
                free_items += 1
            else:
                try:
                    price = float(value)
                    total_items += 1
                    total_amount += price
                except ValueError:
                    print("Invalid price for item:", item)

    final_amount = total_amount - discount

    print("No of items purchased:", total_items)
    print("No of free items:", free_items)
    print("Amount to pay:", int(total_amount) if total_amount.is_integer() else total_amount)
    print("Discount given:", int(discount) if discount.is_integer() else discount)
    print("Final amount paid:", int(final_amount) if final_amount.is_integer() else final_amount)

except FileNotFoundError:
    print("Error: File not found.")

except PermissionError:
    print("Error: Permission denied while opening the file.")

except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
#q2
try:
    num = int(input("Enter a number: "))
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                print(f"{num} is not a prime number.")
                break
        else:
            print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
except ValueError:
    print("Error: Invalid input. Please enter a valid number.")
```

```python
#q4
numbers = [10, -5, 20, -15, 30, 0, 45, -2, 8, -12]

try:
    index = int(input("Enter an index (0-9): "))
    val = numbers[index]

    if val > 0:
        print(f"The number {val} at index {index} is Positive.")
    elif val < 0:
        print(f"The number {val} at index {index} is Negative.")
    else:
        print(f"The number at index {index} is Zero.")

except IndexError:
    print("Error: Invalid index entered. Please stay within range.")
except ValueError:
    print("Error: Please enter a valid integer for the index.")
```
